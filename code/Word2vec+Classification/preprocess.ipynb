# define function

```python
# filtered_df = filtered_temp_df[:15]
# filtered_df = filtered_temp_df
def preprocess_csv(filtered_df):
    idx = 0
    filtered_temp_processed_df = pd.DataFrame(columns = ["case_id", "activity_name", "label"])
    unique_cases = filtered_df['case_id'].unique()
    for case_index, case in enumerate(unique_cases):
        case_id_list = filtered_df[filtered_df['case_id'] == case]['case_id'].to_list()
        case_id_list.extend([case_id_list[0], case_id_list[0]])

        activity_name_list = filtered_df[filtered_df['case_id'] == case]['activity_name'].to_list()
        activity_name_list.insert(0, '▶')
        activity_name_list.insert(len(activity_name_list), '■')

        label_list = filtered_df[filtered_df['case_id'] == case]['label'].to_list()
        label_list.extend([label_list[0], label_list[0]])

        for j in range(len(case_id_list)):
            filtered_temp_processed_df.at[idx, "case_id"]  =  case_id_list[j]
            filtered_temp_processed_df.at[idx, "activity_name"] =  activity_name_list[j]
            if label_list[0] == 'normal':
                filtered_temp_processed_df.at[idx, "label"] = "normal"
            else:
                filtered_temp_processed_df.at[idx, "label"] = "anomalous"
            idx = idx + 1

    return filtered_temp_processed_df
```

```python
original_csv_dir = './injected_csv_for_preprocess/'
destination_csv_dir =  './preprocessed_csv/'
for file in os.listdir(original_csv_dir):
    print(file + '...')
    file_name = file.split('-1.')[0]
    temp_df = pd.read_csv(original_csv_dir + file)
    temp_df = temp_df[['concept:name', 'name', 'label']]
    temp_df.rename(columns={'concept:name':'case_id', 'name':'activity_name'}, inplace = True)
    preprocessed_df = preprocess_csv(temp_df)
    preprocessed_df.to_csv(destination_csv_dir + file_name + '.csv', index=None)
    print(file + ' done')
```
